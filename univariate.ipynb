```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# %%
```

```python
df = pd.read_csv("Iris.csv")

df.head(5)
```

```python
#df.drop(["Id"],axis=1)
df.drop(columns="Id")
```

```python
df.shape
```

```python
df_setosa = df.loc[df["Species"]=="Iris-setosa"]
df_virginica = df.loc[df["Species"]=="Iris-virginica"]
df_vesicolor = df.loc[df["Species"]=="Iris-versicolor"]
```

```python
plt.plot(df_setosa["SepalLengthCm"]),np.zeros_like(df_setosa["SepalLengthCm"])
plt.plot(df_vesicolor["SepalLengthCm"]),np.zeros_like(df_vesicolor["SepalLengthCm"])
plt.plot(df_virginica['SepalLengthCm']),np.zeros_like(df_virginica['SepalLengthCm'])
plt.legend()
plt.xlabel("petal length")
plt.show()
```

```python
###Bivariate Analysis
sns.FacetGrid(df,hue="Species").map(plt.scatter,"SepalLengthCm","SepalWidthCm").add_legend()
```

```python
sns.FacetGrid(df,hue="Species").map(plt.scatter,"PetalLengthCm","PetalWidthCm").add_legend()
```

```python
## Multivariate Analysis
sns.pairplot(df,hue="Species",size=4)
```

```python
sns.jointplot(df)
```
